```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('lab.jpeg')
image_bw = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Equalize the image using histogram equalization
normal_hist = cv2.equalizeHist(image_bw)

# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=6)
final_img = clahe.apply(image_bw)

# Plotting the images and their histograms
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Histogram for Original Image
plt.subplot(2, 3, 4)
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Histogram (Original)')

# Histogram Equalized Image
plt.subplot(2, 3, 2)
plt.imshow(normal_hist, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

# Histogram for Histogram Equalized Image
plt.subplot(2, 3, 5)
plt.hist(normal_hist.ravel(), 256, [0, 256])
plt.title('Histogram (Histogram Equalized)')

# CLAHE Enhanced Image
plt.subplot(2, 3, 3)
plt.imshow(final_img, cmap='gray')
plt.title('CLAHE Enhanced Image')
plt.axis('off')

# Histogram for CLAHE Enhanced Image
plt.subplot(2, 3, 6)
plt.hist(final_img.ravel(), 256, [0, 256])
plt.title('Histogram (CLAHE Enhanced)')

plt.tight_layout()
plt.show()
```
