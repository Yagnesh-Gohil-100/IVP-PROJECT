```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('lena.jpeg')
b, g, r = cv2.split(image)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(image_hsv)
v = cv2.equalizeHist(v)
merged_hsv = cv2.merge((h, s, v))
bgr_enhanced = cv2.cvtColor(merged_hsv, cv2.COLOR_HSV2BGR)

plt.figure(figsize=(12, 8))

# Original Image and Enhanced Image (horizontal layout)
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image (BGR)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(bgr_enhanced, cv2.COLOR_BGR2RGB))
plt.title('Enhanced Image (BGR)')
plt.axis('off')

# Histogram for Original Image (horizontal layout)
plt.subplot(2, 2, 2)

hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

plt.plot(hist_b, color='blue', label='Blue Channel')
plt.plot(hist_g, color='green', label='Green Channel')
plt.plot(hist_r, color='red', label='Red Channel')
plt.title('RGB Channel Histograms (Original)')
plt.legend()

# Histogram for Enhanced Image (horizontal layout)
plt.subplot(2, 2, 4)

b_enhanced, g_enhanced, r_enhanced = cv2.split(bgr_enhanced)

hist_b_enhanced = cv2.calcHist([b_enhanced], [0], None, [256], [0, 256])
hist_g_enhanced = cv2.calcHist([g_enhanced], [0], None, [256], [0, 256])
hist_r_enhanced = cv2.calcHist([r_enhanced], [0], None, [256], [0, 256])

plt.plot(hist_b_enhanced, color='blue', label='Blue Channel')
plt.plot(hist_g_enhanced, color='green', label='Green Channel')
plt.plot(hist_r_enhanced, color='red', label='Red Channel')
plt.title('RGB Channel Histograms (Enhanced)')
plt.legend()

plt.tight_layout(pad=2.0)
plt.show()
```
